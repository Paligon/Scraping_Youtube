Scrape trending YouTube videos (https://www.youtube.com/feed/trending) to get the following data on each movie:
•	URL,
•	title,
•	duration,
•	username,
•	views.
1.	Get the data using Xpath,


```python
def youtube(url):
    response = requests.get(url)
    page = response.content
    tree = html.document_fromstring(page)
    div=tree.xpath("//div[@class='yt-lockup-content']")
    links=div[0].xpath("//h3/a/@href")
    title=div[0].xpath("//h3/a/@title")
    durations=[i.text_content() for i in div[0].xpath("//h3/span")]
    usernames=[i.text_content() for i in div[0].xpath("//div/a[@class='yt-uix-sessionlink       spf-link ']")]
    #views=[i.text_content() for i in div[0].xpath("//ul[@class='yt-lockup-meta-info']/li[2]")]
    views=[i.text_content() for i in div[0].xpath("//li[contains(text(), 'views')]")]
    info=pd.DataFrame({"titles":title,"links":links,"usernames":usernames,"durations":durations,
         "views":views})
    return info
```

```python
url="https://www.youtube.com/feed/trending"
data=youtube(url)
data
```

```python
def make_num(strings):
    """this fuction we will use to convert column 'views' to numeris in order
    to calculate average"""
    strings=int(re.sub('[^0-9,]', "", strings).replace(",",""))
    return strings
def col_range(column):
    """we will use this to calculate calculate range"""
    maxi=column.max()
    mini=column.min()
    result=maxi-mini
    return result
```

```python
data["views"]=data["views"].apply(make_num)## we need this to make numeric
```

```python
c_range=col_range(data["views"])
pprint("The average of views for trending videos is %d"%(data['views'].mean()))
pprint("The median of views for trending videos is %d"%(data['views'].median()))
pprint("The range of views for trending videos is %d"%(c_range))
```

```python
def func(row):
    xml = ['<item>']
    for field in row.index:
        xml.append('  <field name="{0}">{1}</field>'.format(field, row[field]))
    xml.append('</item>')
    return '\n'.join(xml)
```

```python
xml_file='\n'.join(data.apply(func, axis=1))
xml_file
```
